```python
from deepface import DeepFace
import os

TP = FP = TN = FN = 0

def evaluate_folder(folder_path, label):
    global TP, FP, TN, FN
    for pair in os.listdir(folder_path):
        img1 = os.path.join(folder_path, pair, "img1.jpg")
        img2 = os.path.join(folder_path, pair, "img2.jpg")
        try:
            result = DeepFace.verify(img1_path=img1, img2_path=img2, model_name="Facenet512")
            if result["verified"] == label:
                if label: TP += 1
                else: TN += 1
            else:
                if label: FN += 1
                else: FP += 1
        except Exception as e:
            print(f"Error on {pair}: {e}")

evaluate_folder("dataset/positive", True)
evaluate_folder("dataset/negative", False)

total = TP + TN + FP + FN
accuracy = (TP + TN) / total if total else 0
precision = TP / (TP + FP) if (TP + FP) else 0
recall = TP / (TP + FN) if (TP + FN) else 0
f1 = 2 * precision * recall / (precision + recall) if (precision + recall) else 0

print(f"TP: {TP}, FP: {FP}, TN: {TN}, FN: {FN}")
print(f"Accuracy: {accuracy:.2%}")
print(f"Precision: {precision:.2%}")
print(f"Recall: {recall:.2%}")
print(f"F1 Score: {f1:.2%}")
```
